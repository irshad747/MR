Net Promoter Score (NPS). The NPS can be computed using the "Willingness to recommend" scores by classifying responses into Promoters (9-10), Passives (7-8), and Detractors (0-6), and then applying the formula:

NPS
=
(
Number of Promoters
−
Number of Detractors
Total Responses
)
×
100
NPS=(
Total Responses
Number of Promoters−Number of Detractors
​
 )×100

```python
# Defining the NPS calculation function
def calculate_nps(scores):
    promoters = scores.apply(lambda x: 1 if x in [9, 10] else 0).sum()
    detractors = scores.apply(lambda x: 1 if x in range(0, 7) else 0).sum()
    nps = (promoters - detractors) / len(scores) * 100
    return nps

# Calculating NPS for various segments
overall_nps = calculate_nps(cleaned_data['Recommendation_Score'])
nps_by_gender = cleaned_data.groupby('Gender')['Recommendation_Score'].apply(calculate_nps)
nps_by_brand = cleaned_data.groupby('Brand_code')['Recommendation_Score'].apply(calculate_nps)
nps_by_age_group = cleaned_data.groupby('Age Group')['Recommendation_Score'].apply(calculate_nps)

overall_nps
```

Other way to calculate best and worst parametres (Positive-negative deviation from industry average
) i.e  For each brand, the attribute with the highest positive deviation will be considered the "best" parameter, and the one with the highest negative deviation will be considered the "worst" parameter.
Positive Deviation refers to a situation where a brand's performance score on a specific attribute is higher than the overall average score for that attribute across all brands. This indicates that the brand is performing better than the industry standard in that particular area.

Negative Deviation, on the other hand, occurs when a brand's performance score is lower than the overall average. This suggests that the brand is performing worse than the industry standard for that attribute.

```python
# Correcting the method to calculate deviations properly
overall_avg_attributes = cleaned_data[performance_columns].mean()
brand_avg_attributes = cleaned_data.groupby('Brand_code')[performance_columns].mean()
# Calculating deviations from the overall average
brand_deviations = brand_avg_attributes.subtract(overall_avg_attributes, axis=1)

overall_avg_attributes
```

```python
brand_deviations
```

```python
# Identifying the best and worst parameters for each brand
best_parameters = brand_deviations.idxmax(axis=1)
worst_parameters = brand_deviations.idxmin(axis=1)
best_parameters
```

```python
worst_parameters
```

```python
best_worst = pd.DataFrame({
    'Best Parameter': best_parameters,
    'Best Deviation': brand_deviations.max(axis=1),
    'Worst Parameter': worst_parameters,
    'Worst Deviation': brand_deviations.min(axis=1)
})
best_worst
```
